Q1.  What is the difference between AI, ML, DL, and Data Science? Provide a
brief explanation of each?
ANS. Artificial Intelligence (AI): The broadest field, encompassing the development of machines that can perform tasks that typically require human intelligence. This includes things like problem-solving, learning, understanding language, and perception.

Machine Learning (ML): A subset of AI that focuses on enabling systems to learn from data without being explicitly programmed. Instead of hard-coding rules, ML algorithms build models based on data to make predictions or decisions.

Deep Learning (DL): A subfield of Machine Learning inspired by the structure and function of the human brain. It uses artificial neural networks with multiple layers (hence "deep") to learn complex patterns from large amounts of data, often leading to highly accurate models for tasks like image recognition and natural language processing.

Data Science: An interdisciplinary field that uses scientific methods, processes, algorithms, and systems to extract knowledge and insights from structured and unstructured data. It combines elements of statistics, computer science, and domain expertise to analyze data and inform decision-making. Data Science often utilizes AI, ML, and DL techniques as tools in its analytical process.



Q2.  Explain overfitting and underfitting in ML. How can you detect and prevent
them?
ANS. Overfitting occurs when a model learns the training data too well, including the noise and random fluctuations in the data. Instead of learning the general patterns, the model essentially memorizes the training examples. As a result, it performs exceptionally well on the training data but poorly on new, unseen data.
Detection:

High Training Accuracy, Low Test Accuracy: The most common indicator. If your model performs very well on the training set but significantly worse on a separate validation or test set, it's likely overfitting.
Complex Model: Overfitting is often associated with overly complex models that have too many parameters relative to the amount of training data.
Prevention:

More Training Data: Providing more diverse and representative training data can help the model learn general patterns rather than memorizing specific examples.
Feature Selection/Engineering: Reducing the number of features or creating more meaningful features can simplify the model.
Regularization: Techniques like L1 (Lasso) and L2 (Ridge) regularization add a penalty term to the loss function for large coefficients, discouraging the model from becoming too complex and fitting the noise. This effectively shrinks the coefficients, making the model simpler.
Cross-Validation: Techniques like k-fold cross-validation help in getting a more robust estimate of the model's performance on unseen data, allowing for earlier detection of overfitting. It helps in tuning hyperparameters that control model complexity.
Early Stopping: For iterative training algorithms (like neural networks), stop training when the performance on a validation set starts to degrade, even if the training set performance is still improving.
Ensemble Methods: Methods like Random Forests and Gradient Boosting combine multiple models, which can reduce overfitting.
Simpler Model: Choose a less complex model if appropriate for the data.

Underfitting occurs when a model is too simple to capture the underlying patterns in the training data. It fails to learn the relationships between features and the target variable, leading to poor performance on both the training data and new, unseen data.

Detection:

Low Training Accuracy, Low Test Accuracy: If your model performs poorly on both the training set and the test set, it's a strong sign of underfitting.
Simple Model: Underfitting is often associated with models that are too basic for the complexity of the problem.
Prevention:

More Complex Model: Use a more sophisticated model with more parameters or increase the complexity of the existing model (e.g., adding more layers or neurons to a neural network, using a higher degree polynomial in regression).
More Features: Add more relevant features to the dataset so the model has more information to learn from.
Reduce Regularization: If regularization is being used, reducing its strength can allow the model to fit the data more closely.
Feature Engineering: Create new features that might better capture the underlying relationships in the data.
Increase Training Time: For iterative algorithms, training for a longer duration might allow the model to learn more from the data.

Bias-Variance Tradeoff
The concepts of overfitting and underfitting are closely related to the bias-variance tradeoff.

Bias: This refers to the error introduced by approximating a real-world problem, which may be complex, by a simplified model. A high bias model makes strong assumptions about the data, leading to underfitting. It consistently misses the mark because it can't capture the underlying patterns.
Variance: This refers to the amount that the estimate of the target function will change if different training data was used. A high variance model is very sensitive to the training data, leading to overfitting. It learns the noise in the training data and performs poorly on unseen data because it's too specific to the training set.
The goal in machine learning is to find a balance between bias and variance. A model with high bias (underfit) will have low variance, and a model with high variance (overfit) will have low bias. The optimal model has both low bias and low variance, meaning it generalizes well to new data. Techniques like regularization and cross-validation are used to navigate this tradeoff and find the sweet spot for model complexity.

Q3.  How would you handle missing values in a dataset? Explain at least three
methods with examples.
ANS. andling missing values is a crucial step in data preprocessing, as missing data can lead to biased results or errors in machine learning models. Here are three common methods for handling missing values, with explanations and examples:

1. Imputation (Replacing Missing Values)
Imputation involves filling in the missing values with estimated values. The choice of imputation strategy depends on the nature of the data and the extent of missingness.

a) Mean/Median Imputation:

Explanation: This method replaces missing numerical values with the mean or median of the non-missing values in that column. Mean imputation is sensitive to outliers, while median imputation is more robust.
When to Use: Suitable for numerical features where the missing values are assumed to be randomly distributed, and the distribution of the non-missing data is roughly symmetrical (for mean) or skewed (for median).
Example (Python with Pandas):
import pandas as pd
import numpy as np

data = {'A': [1, 2, np.nan, 4, 5],
        'B': [10, np.nan, 30, 40, 50]}
df = pd.DataFrame(data)

# Impute missing values in column 'A' with its mean
df['A'].fillna(df['A'].mean(), inplace=True)

# Impute missing values in column 'B' with its median
df['B'].fillna(df['B'].median(), inplace=True)

print(df)
b) Mode Imputation:

Explanation: This method replaces missing categorical values (or sometimes numerical values) with the most frequent value (mode) in that column.
When to Use: Best for categorical features or numerical features with a clear mode, where the missing values are likely to belong to the most common category.
Example (Python with Pandas):
import pandas as pd
import numpy as np

data = {'Category': ['A', 'B', np.nan, 'A', 'C'],
        'Color': ['Red', 'Blue', 'Red', np.nan, 'Green']}
df = pd.DataFrame(data)

# Impute missing values in 'Category' with its mode
df['Category'].fillna(df['Category'].mode()[0], inplace=True)

# Impute missing values in 'Color' with its mode
df['Color'].fillna(df['Color'].mode()[0], inplace=True)

print(df)
c) Forward Fill (ffill) / Backward Fill (bfill):

Explanation: These methods are typically used for time-series or ordered data. Forward fill propagates the last valid observation forward to the next missing value, while backward fill propagates the next valid observation backward.
When to Use: Suitable for time-series data where the value at a certain point is likely to be similar to the preceding or succeeding value.
Example (Python with Pandas):
import pandas as pd
import numpy as np

data = {'Date': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']),
        'Sales': [100, np.nan, 120, np.nan, 150]}
df = pd.DataFrame(data)

# Forward fill missing sales values
df['Sales_ffill'] = df['Sales'].fillna(method='ffill')

# Backward fill missing sales values
df['Sales_bfill'] = df['Sales'].fillna(method='bfill')

print(df)
2. Deletion
Deletion involves removing rows or columns that contain missing values.

a) Row-wise Deletion (Listwise Deletion):

Explanation: If a row contains any missing value, the entire row is removed from the dataset.
When to Use: Only advisable when the number of missing values is very small relative to the dataset size, and dropping rows won't significantly reduce the dataset's representativeness. If many rows have missing values, this can lead to substantial data loss and biased results.
Example (Python with Pandas):
import pandas as pd
import numpy as np

data = {'A': [1, 2, np.nan, 4, 5],
        'B': [10, 20, 30, np.nan, 50]}
df = pd.DataFrame(data)

df_cleaned = df.dropna()
print(df_cleaned)
b) Column-wise Deletion:

Explanation: If a column has too many missing values (e.g., above a certain threshold like 70% or 80%), the entire column is removed.
When to Use: When a column has an overwhelming proportion of missing data, making it unreliable for analysis or modeling. It's often better to drop such a column than to impute it extensively, as imputation might introduce too much artificial information.
Example (Python with Pandas):
import pandas as pd
import numpy as np

data = {'A': [1, 2, 3, 4, 5],
        'B': [10, 20, np.nan, np.nan, 50],
        'C': [np.nan, np.nan, np.nan, np.nan, 100]}
df = pd.DataFrame(data)

# Drop columns where more than 50% of values are missing
threshold = len(df) * 0.5
df_cleaned = df.dropna(axis=1, thresh=threshold)
print(df_cleaned)
3. Advanced Imputation Techniques
For more complex scenarios, especially with larger datasets and patterns in missingness, advanced techniques can be employed.

a) Regression Imputation:

Explanation: This method uses a regression model to predict the missing values based on other features in the dataset. For example, if 'Age' is missing, you could build a regression model using 'Education', 'Income', etc., to predict the missing 'Age' values.
When to Use: When there's a strong correlation between the feature with missing values and other features in the dataset. It's more sophisticated than simple mean/median/mode imputation but also more computationally intensive.
b) K-Nearest Neighbors (KNN) Imputation:

Explanation: This method imputes missing values by using the values from the k nearest neighbors. For a missing value in a specific row, KNN finds the k most similar rows (neighbors) based on other features and uses their values to estimate the missing one (e.g., taking the mean for numerical or mode for categorical).
When to Use: Effective when there are patterns in the data that can be captured by proximity, and it can handle both numerical and categorical data. It can be computationally expensive for very large datasets.

Q4.  What is an imbalanced dataset? Describe two techniques to handle it
ANS. An imbalanced dataset refers to a classification problem where the distribution of examples across the known classes is skewed. That is, one class (the majority class) significantly outnumbers the other class(es) (the minority class). This imbalance can lead to machine learning models that perform well on the majority class but poorly on the minority class, as the model may optimize for overall accuracy by simply predicting the majority class.

For example, in fraud detection, legitimate transactions (majority class) might be 99.9% of the data, while fraudulent transactions (minority class) are only 0.1%. A model that always predicts "legitimate" would have 99.9% accuracy but fail to detect any fraud.

Here are two common techniques to handle imbalanced datasets:

1. Resampling Techniques
Resampling techniques aim to change the distribution of the dataset to balance the class proportions. These can be broadly categorized into oversampling and undersampling.

a) Oversampling the Minority Class:

Explanation: This technique involves increasing the number of instances in the minority class to match the number of instances in the majority class. Common methods include:
Random Oversampling: Duplicating random instances from the minority class.
SMOTE (Synthetic Minority Over-sampling Technique): This is a more sophisticated method that generates synthetic samples for the minority class. It works by selecting a minority class instance and then finding its K-nearest neighbors. It then creates new synthetic instances along the line segments joining the minority instance and its selected neighbors.
When to Use: When the dataset size is not very large, and you want to ensure the model has enough examples of the minority class to learn from. SMOTE is generally preferred over random oversampling as it introduces new information.
Example (Conceptual with SMOTE): Imagine a dataset with 100 fraudulent transactions and 9900 legitimate transactions. SMOTE would generate new synthetic fraudulent transactions until the number of fraudulent transactions is closer to (or equal to) 9900.
b) Undersampling the Majority Class:

Explanation: This technique involves reducing the number of instances in the majority class to balance the class distribution. Common methods include:
Random Undersampling: Randomly removing instances from the majority class.
Tomek Links: Identifies pairs of instances from different classes that are each other's nearest neighbors. The majority class instance from such a pair is then removed.
Edited Nearest Neighbors (ENN): Removes majority class instances whose class label is different from the class label of at least two of its three nearest neighbors.
When to Use: When the dataset is very large, and you want to reduce the training time. However, a major drawback is the potential loss of valuable information from the majority class.
Example (Conceptual with Random Undersampling): From the fraud detection example, you would randomly remove 9800 legitimate transactions, leaving you with 100 fraudulent and 100 legitimate transactions.
2. Algorithmic Approaches / Cost-Sensitive Learning
Instead of modifying the dataset, these techniques involve using algorithms that are designed to work well with imbalanced data, or modifying existing algorithms to be sensitive to the class imbalance.

a) Use of Tree-Based Algorithms (e.g., Random Forest, Gradient Boosting):

Explanation: Ensemble methods like Random Forest often perform well on imbalanced datasets without explicit resampling. This is because they combine predictions from multiple trees, and individual trees can focus on different aspects of the data. Random Forest, for instance, can sometimes balance the errors from the majority and minority classes naturally.
When to Use: Often a good first approach, as these algorithms are robust and powerful. They can capture complex patterns and are less sensitive to noise.
Example: Training a Random Forest classifier directly on the imbalanced fraud detection dataset. The ensemble nature helps in preventing the model from being overly biased towards the majority class.
b) Cost-Sensitive Learning (Adjusting Class Weights):

Explanation: Many machine learning algorithms (like Support Vector Machines, Logistic Regression, and neural networks) allow you to assign different weights to different classes during training. By assigning a higher penalty (cost) to misclassifying the minority class and a lower penalty to misclassifying the majority class, the model is encouraged to pay more attention to the minority class.
When to Use: When the cost of misclassifying the minority class is much higher than misclassifying the majority class (e.g., failing to detect fraud is more costly than a false alarm). This approach doesn't alter the data distribution but influences the learning process.
Example: In scikit-learn, many classifiers have a class_weight parameter (e.g., class_weight='balanced' or class_weight={0: 1, 1: 10}). Setting class_weight='balanced' automatically adjusts weights inversely proportional to class frequencies, giving more weight to the minority class. If class 0 (legitimate) is 99% and class 1 (fraud) is 1%, you might set class_weight={0: 1, 1: 99}

Q5 Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.  
ANS. Feature scaling is crucial for several reasons, primarily because many machine learning algorithms are sensitive to the scale of input features:

Gradient Descent-Based Algorithms: Algorithms like Linear Regression, Logistic Regression, Neural Networks, and Support Vector Machines (SVMs) use gradient descent to optimize their cost function. If features are on different scales, the cost function will have an elongated, narrow shape, making the optimization process slower and more prone to getting stuck in local minima. Scaling ensures that all features contribute equally to the gradient, leading to faster and more stable convergence.

Distance-Based Algorithms: Algorithms such as K-Nearest Neighbors (KNN), K-Means Clustering, and SVMs (especially with radial basis function kernels) rely on calculating distances between data points. If one feature has a much larger range than others, it will disproportionately influence the distance calculation, effectively overshadowing the contributions of other features. Scaling ensures that all features contribute equally to the distance metric.

Regularization Techniques: Regularization methods (L1, L2) penalize large coefficients to prevent overfitting. If features are not scaled, features with larger values might end up with smaller coefficients just to balance their magnitude, which can lead to unfair penalization. Scaling ensures that the regularization penalty is applied fairly across all features.

Interpretability of Coefficients: In models like linear regression, if features are scaled, the magnitude of their coefficients can be more directly compared to understand their relative importance. Without scaling, a small coefficient for a feature with a large range might actually indicate a strong impact, while a large coefficient for a feature with a small range might indicate a weak impact.

Comparison: Min-Max Scaling vs. Standardization
These are two of the most common feature scaling techniques, each with its own characteristics and use cases.

1. Min-Max Scaling (Normalization)
Formula: (X_{scaled} = \frac{X - X_{min}}{X_{max} - X_{min}})

Concept: This technique rescales the features to a fixed range, typically between 0 and 1 (or -1 and 1 if specified). It shifts and rescales values so that they are bound within a specific interval.

Characteristics:

Range: Transforms data to a specific range (e.g., [0, 1]).
Distribution: Does not change the shape of the original distribution.
Outlier Sensitivity: Highly sensitive to outliers. A single outlier can significantly compress the range of other data points, making them very close to each other.
Advantages:

Simple and intuitive to understand.
Preserves the exact relationships between data points.
Useful when algorithms require features to be within a specific bounded range (e.g., some neural network activation functions expecting input between 0 and 1).
Disadvantages:

Sensitive to outliers. If new, unseen data contains values outside the original min or max, the scaled values will also fall outside the [0, 1] range, requiring re-scaling.
Loss of information about the original magnitude if the range is very small due to outliers.
2. Standardization (Z-score Normalization)
Formula: (X_{scaled} = \frac{X - \mu}{\sigma}) where (\mu) is the mean of the feature and (\sigma) is its standard deviation.

Concept: This technique transforms the data to have a mean of 0 and a standard deviation of 1. It centers the data around the mean and scales it by the standard deviation.

Characteristics:

Range: Does not bound values to a specific range. Values can be anywhere from negative to positive infinity, though most will lie between -3 and 3 for normally distributed data.
Distribution: Results in a distribution with a mean of 0 and a standard deviation of 1. It is particularly useful if the data follows a Gaussian (normal) distribution, but it doesn't make non-normal data normal.
Outlier Sensitivity: Less affected by outliers than Min-Max scaling because it uses the mean and standard deviation, which are more robust to extreme values than the absolute min/max.
Advantages:

Less sensitive to outliers compared to Min-Max scaling.
Useful for algorithms that assume a Gaussian distribution of features (e.g., Linear Regression, Logistic Regression, Gaussian Naive Bayes).
Does not compress the range of data points in the presence of outliers as much as Min-Max scaling.
It's generally preferred when the algorithm requires features to have a mean of 0.
Disadvantages:

Does not guarantee a specific output range.
Can be less intuitive to interpret the scaled values compared to Min-Max scaling.
When to Choose Which?
Choose Min-Max Scaling:

When the data is not expected to have extreme outliers.
When the algorithm requires features to be in a specific, bounded range (e.g., neural networks with sigmoid activation functions).
When you need to preserve the exact original relationships of values within the feature.
Choose Standardization:

When the data has outliers, or you suspect it might.
When the algorithm assumes a normal distribution (or is sensitive to scale but not to a bounded range).
When distance-based algorithms are used, and you want each feature to contribute equally to the distance calculation without being overly influenced by individual feature ranges.
It's often a good default choice when you're unsure, as it performs well for a wide range of algorithms.
In summary, both techniques aim to bring features to a comparable scale, but they achieve this differently, with implications for their robustness to outliers and suitability for various machine learning algorithms.



Q6.  Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?
ANS. Label Encoding and One-Hot Encoding are two common techniques used to convert categorical data into numerical format, which is essential for most machine learning algorithms. However, they have distinct applications and implications.

Label Encoding
Concept: Label Encoding assigns a unique numerical integer to each category in a feature. For example, if a feature 'Color' has categories 'Red', 'Green', 'Blue', they might be encoded as 0, 1, 2 respectively.

Characteristics:

Output: Single numerical column.
Ordinality: Introduces an artificial ordinal relationship (order) between categories. For instance, in the 'Color' example, 'Blue' (2) is treated as 'greater' than 'Green' (1), which is 'greater' than 'Red' (0).
Dimensionality: Does not increase the dimensionality of the dataset.
When to Prefer Label Encoding:

Ordinal Features: When the categorical feature inherently has an order or rank. For example, 'Education Level' (High School, Bachelor's, Master's, PhD) can be encoded as 0, 1, 2, 3.
Tree-Based Algorithms: Algorithms like Decision Trees, Random Forests, and Gradient Boosting Machines (e.g., XGBoost, LightGBM) can sometimes handle the ordinal relationship introduced by Label Encoding without issues. These algorithms make decisions based on splits, and the exact numerical magnitude between categories is less critical than with linear models.
High Cardinality: When you have a very large number of unique categories in a feature (high cardinality) and using One-Hot Encoding would lead to an excessively high number of new features, potentially causing the 'curse of dimensionality' or making the model computationally expensive.
Disadvantages:

Machine learning algorithms might interpret the assigned numerical values as having a magnitude or order, even when there is none (nominal data). This can lead to incorrect assumptions and poorer performance, especially with linear models.
One-Hot Encoding
Concept: One-Hot Encoding converts each category into a new binary feature (column) and assigns a 1 to the column corresponding to the category and 0 to all other category columns. For the 'Color' example, 'Red' would become [1, 0, 0], 'Green' would be [0, 1, 0], and 'Blue' would be [0, 0, 1].

Characteristics:

Output: Creates multiple new binary columns (one for each unique category, minus one if dropping the first to avoid multicollinearity).
Ordinality: Avoids implying any ordinal relationship between categories.
Dimensionality: Increases the dimensionality of the dataset (number of features).
When to Prefer One-Hot Encoding:

Nominal Features: When the categorical feature has no inherent order or rank (i.e., nominal data). Most categorical features fall into this category (e.g., 'City', 'Marital Status', 'Product Type').
Linear Models and Distance-Based Algorithms: Algorithms such as Linear Regression, Logistic Regression, Support Vector Machines (SVMs), K-Nearest Neighbors (KNN), and Neural Networks perform much better with One-Hot Encoded features. These algorithms are sensitive to the magnitude and scale of features, and an artificial ordering introduced by Label Encoding can lead to biased results.
Low to Moderate Cardinality: When the number of unique categories is small to moderate, such that the creation of new columns does not excessively increase the dataset's dimensionality.
Disadvantages:

Increased Dimensionality: Can lead to a very sparse dataset with many zero values and significantly increase the number of features, especially with high cardinality features. This can increase computational cost and potentially lead to the 'curse of dimensionality', making models harder to train and interpret.
Multicollinearity: If not handled carefully (e.g., by dropping one of the generated columns), it can introduce multicollinearity, which can be an issue for some models like linear regression.
Summary of Choice:
Use One-Hot Encoding for nominal categorical data and with linear or distance-based models (e.g., Logistic Regression, SVM, KNN).
Use Label Encoding for ordinal categorical data or when working with tree-based models (e.g., Decision Trees, Random Forests, Gradient Boosting) where the introduced ordering might not negatively impact performance, or when dealing with very high cardinality to avoid excessive dimensionality.

Q7. Analyze the relationship between app categories and ratings. Which categories have the
highest/lowest average ratings, and what could be the possible reasons?

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
df.dropna(subset=['Rating'], inplace=True)
print(f"Shape of DataFrame after dropping rows with missing ratings: {df.shape}")
```

```python
df['Rating'] = pd.to_numeric(df['Rating'])
print(f"Data type of 'Rating' column after conversion: {df['Rating'].dtype}")
```

```python
unique_categories = df['Category'].unique()
print(f"Number of unique categories: {len(unique_categories)}")
print("Unique categories:\n", unique_categories)
```

```python
df = df[df['Category'] != '1.9']
print(f"Shape of DataFrame after removing '1.9' category: {df.shape}")
print(f"Unique categories after cleaning: {df['Category'].unique().tolist()}")
```

```python
avg_ratings_by_category = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
print("Average ratings by category:\n", avg_ratings_by_category.head())
```

```python
highest_rated_category = avg_ratings_by_category.index[0]
highest_avg_rating = avg_ratings_by_category.iloc[0]
lowest_rated_category = avg_ratings_by_category.index[-1]
lowest_avg_rating = avg_ratings_by_category.iloc[-1]

print(f"Category with the highest average rating: '{highest_rated_category}' (Rating: {highest_avg_rating:.2f})")
print(f"Category with the lowest average rating: '{lowest_rated_category}' (Rating: {lowest_avg_rating:.2f})")
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x=avg_ratings_by_category.index, y=avg_ratings_by_category.values, palette='viridis')
plt.title('Average App Ratings by Category', fontsize=16)
plt.xlabel('App Category', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

Q10(A)  What factors most strongly correlate with employee attrition? Use visualizations to show key
drivers (e.g., satisfaction, overtime, salary).

```python
df_hr = pd.read_csv('hr_analytics.csv')
print("HR Analytics dataset loaded successfully.")
print(df_hr.head())
```

```python
print(f"Data type of 'left' column: {df_hr['left'].dtype}")
missing_left = df_hr['left'].isnull().sum()
print(f"Missing values in 'left' column: {missing_left}")

# If there are missing values, drop them
if missing_left > 0:
    df_hr.dropna(subset=['left'], inplace=True)
    print(f"DataFrame shape after dropping rows with missing 'left' values: {df_hr.shape}")
```

```python
salary_mapping = {'low': 0, 'medium': 1, 'high': 2}
df_hr['salary_numerical'] = df_hr['salary'].map(salary_mapping)

print("Original 'salary' column unique values:", df_hr['salary'].unique())
print("Converted 'salary_numerical' column unique values:", df_hr['salary_numerical'].unique())
print("First 5 rows with original and converted salary:")
print(df_hr[['salary', 'salary_numerical']].head())
```

```python
numerical_cols_for_correlation = [
    'satisfaction_level',
    'last_evaluation',
    'number_project',
    'average_montly_hours',
    'time_spend_company',
    'Work_accident',
    'promotion_last_5years',
    'salary_numerical'
]

# Check for missing values in the selected numerical columns
missing_values_in_selected_cols = df_hr[numerical_cols_for_correlation].isnull().sum()
print("Missing values in selected numerical columns:\n", missing_values_in_selected_cols[missing_values_in_selected_cols > 0])

# Drop rows with any missing values in these specific columns if any are found
if missing_values_in_selected_cols.sum() > 0:
    df_hr.dropna(subset=numerical_cols_for_correlation, inplace=True)
    print(f"DataFrame shape after dropping rows with missing values in selected columns: {df_hr.shape}")

print("Selected numerical columns for correlation:", numerical_cols_for_correlation)
```

```python
correlations = df_hr[numerical_cols_for_correlation + ['left']].corr()['left'].drop('left').sort_values(ascending=False)
print("Correlation with 'left' (Attrition) column:")
print(correlations)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=correlations.index, y=correlations.values, palette='coolwarm')
plt.title('Correlation of Features with Employee Attrition (Left)', fontsize=16)
plt.xlabel('Feature', fontsize=12)
plt.ylabel('Correlation Coefficient', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
